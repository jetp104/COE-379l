```python
duplicate_rows = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
data = data.drop_duplicates()
```

```python
missing_values = data.isnull().sum()
print("Missing values per column:\n", missing_values)
```

```python
columns_to_analyze = [
    "MedInc", "HouseAge", "AveRooms", "AveBedrms", "Population",
    "AveOccup", "Latitude", "Longitude", "price_above_median"
]

# Compute statistics
stats = {
    "Mean": data[columns_to_analyze].mean(),
    "Median": data[columns_to_analyze].median(),
    "Std Dev": data[columns_to_analyze].std(),
    "Variance": data[columns_to_analyze].var()
}

stats_df = pd.DataFrame(stats)

print(stats_df)

print("\nModes for each column:")
for col in columns_to_analyze:
    modes = data[col].mode().tolist()  # Convert mode(s) to a list
    print(f"{col}: {modes}")
```

```python
#The dataset shows a right-skewed income distribution, with a mean income of 3.87 and a median of 3.53, suggesting some high-income outliers. 
#The mode values (3.125 and 15.0001) indicate two distinct income clusters. Housing age is fairly balanced, with a mean of 28.64 years, but the most 
#common house age is 52, implying many older homes. The average number of rooms is 5.43, with most homes having around 5 rooms, and the typical number of 
#bedrooms is 1. Population per neighborhood varies widely, with an average of 1,425 but a mode of 891, showing some high-density areas. Average occupancy 
#is 3.07 per household, but the high standard deviation (10.39) suggests significant variation. The geographical distribution centers around Southern 
#California, particularly near Los Angeles (latitude ~34.06, longitude ~-118.31). The dataset is evenly split in terms of home prices above and below the 
#median, ensuring a balanced representation.
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

X = data.drop('price_above_median', axis=1)
y = data['price_above_median']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=1)

standard_scaler = StandardScaler()

X_train_standard = standard_scaler.fit_transform(X_train)

X_test_standard = standard_scaler.transform(X_test)

X_train_standard_df = pd.DataFrame(X_train_standard, columns=X_train.columns)
X_test_standard_df = pd.DataFrame(X_test_standard, columns=X_test.columns)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knn = KNeighborsClassifier()

param_grid = {
    "n_neighbors": np.arange(1, 40, 2), 
    "weights": ["uniform", "distance"],
    "metric": ["euclidean", "manhattan"]
}

grid_search = GridSearchCV(knn, param_grid, cv=5, scoring="accuracy", n_jobs=-1)
grid_search.fit(X_train_standard, y_train)

best_knn = grid_search.best_estimator_

y_train_pred = best_knn.predict(X_train_standard)
y_test_pred = best_knn.predict(X_test_standard)

print("Best Parameters:", grid_search.best_params_)
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Classification Report:\n", classification_report(y_test, y_test_pred))
print(f"Performance on TRAIN\n********************\n{classification_report(y_train,y_train_pred)}")

cm_train = confusion_matrix(y_train, y_train_pred)

cm_test = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Training Set (K-Nearest Neighbor)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Test Set (K-Nearest Neighbor)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=1).fit(X_train_standard_df, y_train)

y_train_pred = model.predict(X_train_standard_df)

y_test_pred = model.predict(X_test_standard_df)
```

```python
print(f"Performance on TEST\n*******************\n{classification_report(y_test, y_test_pred)}")
print(f"Performance on TRAIN\n********************\n{classification_report(y_train,y_train_pred)}")
print(f"Test Accuracy: {accuracy_score(y_test, y_test_pred)}\n")
```

```python
cm_train = confusion_matrix(y_train, y_train_pred)

cm_test = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Training Set (Decision Tree)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Test Set (Decision Tree)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=1)

param_grid = {
    'n_estimators': np.arange(1, 100, 2),
    'max_depth': [None] + list(np.arange(2, 10, 2)),  
    'min_samples_split': np.arange(2, 8, 2),  
    'min_samples_leaf': np.arange(1, 5, 1)
}

grid_search = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1, scoring="accuracy")
grid_search.fit(X_train_standard_df, y_train)

best_rf = grid_search.best_estimator_

y_train_pred = best_rf.predict(X_train_standard_df)
y_test_pred = best_rf.predict(X_test_standard_df)

print("Best Hyperparameters:", grid_search.best_params_)
print("Train Accuracy:", accuracy_score(y_train, y_train_pred))
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))

print(f"Performance on TEST\n*******************\n{classification_report(y_test, y_test_pred)}")
print(f"Performance on TRAIN\n********************\n{classification_report(y_train,y_train_pred)}")

cm_train = confusion_matrix(y_train, y_train_pred)
cm_test = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Training Set (Random Forest)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Test Set (Random Forest)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import AdaBoostClassifier

decision_tree_stump = DecisionTreeClassifier(max_depth=6)

param_grid = {
    'learning_rate': [0.1, 0.5, 1.0],
    'n_estimators': [50, 100, 200]
}

abc = AdaBoostClassifier(estimator=decision_tree_stump, random_state=1)

grid_search = GridSearchCV(abc, param_grid, cv=3, n_jobs=8, scoring="accuracy")
grid_search.fit(X_train_standard_df, y_train)

best_abc = grid_search.best_estimator_

y_train_pred = best_abc.predict(X_train_standard_df)

y_test_pred = best_abc.predict(X_test_standard_df)

test_report = classification_report(y_test, y_test_pred)
train_report = classification_report(y_train, y_train_pred)
print("Best Hyperparameters:", grid_search.best_params_)
print(f"Performance on TEST\n*******************\n{test_report}")
print(f"Performance on TRAIN\n********************\n{train_report}")
print(f"Test Accuracy: {accuracy_score(y_test, y_test_pred)}\n")

cm_train = confusion_matrix(y_train, y_train_pred)

cm_test = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Training Set (AdaBoost)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.subplot(1, 2, 2)
sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title('Confusion Matrix - Test Set (AdaBoost)')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```
